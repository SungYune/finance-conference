```python
import pandas as pd
import matplotlib.pyplot as plt

% cd../../
```

```python
dow = pd.read_csv('data/forex/dow.csv')
exchange_rate = pd.read_csv('data/forex/exchange_rate.csv')
fund_rate = pd.read_csv('data/forex/fund_rate.csv')
nasdaq = pd.read_csv('data/forex/nasdaq.csv')
sp500 = pd.read_csv('data/forex/sp500.csv')
t_note = pd.read_csv('data/forex/t_note.csv')
```

```python
def visualize(df, fig, title, i):
    fig.add_subplot(3, 3, i)
    plt.plot(df['date'], df['value'])
    plt.title(title)
    plt.xlabel('date')
    plt.ylabel('value')
    print(i, " done")
```

```python
fig = plt.figure(figsize=(4, 2))
plt.plot(dow['date'], dow['value'])
plt.title('Dow Jones Industrial Average')
plt.xlabel('date')
plt.ylabel('value')
plt.tick_params(
    axis='x',
    which='both',
    bottom=False,
    top=False,
    labelbottom=False)
plt.show()
```
